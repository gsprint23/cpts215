### 2 Hashing
(Chapter 3 in zyBooks; Chapter 5.5 in [Miller and Ranum](http://interactivepython.org/runestone/static/pythonds/index.html))

    * Fast, possible $\mathcal{O}(1)$ data placement and retrieval
* When is hashing appropriate?
    * Don't care about the order of the data
    * Alternative to sorted binary search
* What is a hash function?
    * Must produce the same hash for the same data
    * Index must fit within the table
    * Provide a unique index
    * Provide examples of hash functions
* What is a collision?
    * How to handle a collision?
        * Linear probing
        * Chaining (linked list processing)
* What kind of things can you do to make hashing better?
    * Larger hash table
    * Prime size of hash table
    * Don't allow for zeros in hash function multiplication
* Given a hash function and data, place the data in a hash table
