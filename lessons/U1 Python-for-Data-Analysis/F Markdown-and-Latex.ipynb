## Markdown Overview
Before talking about Markdown, let's talk about [HTML](https://en.wikipedia.org/wiki/HTML) (hypertext markup language). HTML is used to build content for websites. HTML is characterized by tags, which are angle brackets < > used to "label" content. You have probably seen HTML before, maybe something like: 

`<h1>This is a header</h1>`

In the above example, the text "This is a header" is labeled as a first (top) level header. If you want to learn more about HTML (and CSS and Javascript), I recommend starting with [this Codeacademy tutorial](https://www.codecademy.com/learn/web).

Now, let's talk about Markdown. Like HTML, Markdown is used to build content for websites; however, Markdown is much easier to learn and simpler to read. For example, let's look at the previous HTML header example in Markdown:

`# This is a header`

The # symbol denotes a header, and the number of # symbols denotes a header level. Now let's dive into the Markdown (and a little bit of HTML) you will need for this class.

Note: You have likely seen Markdown in use before. Github and Reddit use it to format comments. In fact, [README.md](https://help.github.com/articles/about-readmes/) files on Github are Markdown files (md is for markdown).

Note: **In addition to following along in these notes, I highly recommend following [this Markdown tutorial](http://www.markdowntutorial.com/)** for extra practice and to learn more above and beyond what is presented here.

### Headers
Headers are used to label a section. There are 6 different levels of headers:
# Header Level 1
## Header Level 2
### Header Level 3
#### Header Level 4
##### Header Level 5
###### Header Level 6
Each header (except for level 1) represents a sub section of previous, higher level headers. A phrase is specified as a header by beginning with the # symbol. The number of # symbols denotes the level of the header. For example, this cell is under the `### Headers` header, which is a level 3 header.

### Paragraphs
Newlines in Markdown are handled a little differently than you might think. For example,
```
Line1
Line2
Line3
```
will render as:

Line1
Line2
Line3

To get the lines on their own line, explicitly place a blank line between them (this is called a *hard* break):
```
Line1

Line2

Line3
```
will render as:

Line1

Line2

Line3

Usually we want the lines "tighter" together than this. To get the lines on their own line, without such a large blank space between them, place two spaces at the end of the line (this is called a *soft* break):
```
Line1  
Line2  
Line3  
```
will render as:

Line1  
Line2  
Line3

It is hard to see the double spaces after Line1 and Line2, but if you move your cursor there you will see them!

### Formatting Text
Italics: surround text to be italicized with either `*` or `_`
* Example: `_italicized_` will be rendered as _italicized_
* Example `*italicized*` will be rendered as *italicized*

Bold: surround text to be bolded with `**`
* Example: `**bolded**` will be rendered as **bolded**

Strike through: surround text to be strike through with `~~`
* Example: `~~strike~~` will be rendered as ~~strike~~

Block quotes: preface text to be indented as a block quote with `>`
* Example: `>"Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum."` will be rendered as
>"Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum."

Inline code: surround text to look like code with \`
* Example: \`print(5)\` will be rendered as `print(5)`

Block code: surround text to look like code with \`\`\`
* Example:

\`\`\`
x = 5
for i in range(x):
    print(i, end=", ")
\`\`\`

will be rendered as
```
x = 5
for i in range(x):
    print(i, end=", ")
```

Block code syntax highlighted: same as block code, but add the name of the language after the first \`\`\`
* Example:

\`\`\`python
x = 5
for i in range(x):
    print(i, end=", ")
\`\`\`

will be rendered as
```python
x = 5
for i in range(x):
    print(i, end=", ")
```


### Lists
In Markdown we can have lists with numbers (ordered lists) and lists with bullets (unordered lists).
#### Unordered Lists
Preface each item in an unordered list with a `*`. Place each item on its own line. For example:
```
* Item1
* Item2
* Item3
```
will be rendered as:
* Item1
* Item2
* Item3

#### Ordered Lists
Preface each item in an ordered list with a `1.`. Place each item on its own line. For example:
```
1. Item1
1. Item2
1. Item3
```
will be rendered as:
1. Item1
1. Item2
1. Item3

#### Nested Lists
To nest lists (a list with in a list), we simply need to indent the items using either a space or a tab. For example:
```
1. Item1
    1. SubItem1
    1. SubItem2
        1. SubItem3
1. Item2
1. Item3
    1. SubItem4
```
would be rendered as:
1. Item1
    1. SubItem1
    1. SubItem2
        1. SubItem3
1. Item2
1. Item3
    1. SubItem4
    
And of course, we can nest ordered and unordered lists!
```
1. Item1
    *. SubItem1
    *. SubItem2
1. Item2
    1. SubItem3
    1. SubItem4
```
would be rendered as:
1. Item1
    * SubItem1
    * SubItem2
1. Item2
    1. SubItem3
    1. SubItem4
    
Lastly, we can have paragraphs as part of the "body" of list items. The paragraphs need to be indented by at least once space, and be on a line by itself under the list item. For example (note that I am using double spaces to tighten the text):
```
1. Item1  
   Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua.  
   Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. 
1. Item2  
   Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.
```
would be rendered as:
1. Item1  
   Lorem ipsum dolor sit amet, consectetur adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua.  
   Ut enim ad minim veniam, quis nostrud exercitation ullamco laboris nisi ut aliquip ex ea commodo consequat. 
1. Item2  
   Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum.

### Links
Often we want to link to a other websites. We can do this with links. Links are in the form: `[text label](url)` For example, to link the text "WSU Blackboard" to the WSU Blackboard website, "https://learn.wsu.edu/webapps/login/":

`[WSU Blackboard](https://learn.wsu.edu/webapps/login/)` will be rendered as [WSU Blackboard](https://learn.wsu.edu/webapps/login/)

### Tables
To make a table in Markdown, organize your table items and divide them with - (hyphens) and | (pipes). Hyphens separate the the header row (first row) from the rest of the rows. Rows other than the header row are separated by newlines. Pipes separate items in a row (i.e. columns). For example:

```
First Header | Second Header
------------ | -------------
Content from cell 1 | Content from cell 2
Content in the first column | Content in the second column
```

will render as:

First Header | Second Header
------------ | -------------
Content from cell 1 | Content from cell 2
Content in the first column | Content in the second column

## Latex Overview
[Latex](https://www.latex-project.org/about/) is a markup language to typeset documents. Specifically, Latex is really good at typesetting *math*. We are going to use Latex in our Markdown cells to render beautiful mathematical formulas. How we are going to use Latex only scratches the surface of what Latex can be used for (i.e. writing entire papers, as an alternative to Microsoft Word). 

### Math Mode
To enter Latex code in a Markdown cell, we need to enter "math mode", which is prefacing our Latex code with the \$ symbol for inline Latex, and \$\$ for block Latex. For example:
```
The word pi in Latex, $pi$, looks more "mathy".
```

would render as:

The word pi in Latex, $pi$, looks more "mathy".

And block Latex:
```
The word pi in Latex
$$pi$$
is now centered and on its own line.
```

would render as:

The word pi in Latex

$$pi$$

is now centered and on its own line.

Now let's learn some basic Latex!

### Basic Functionality
Subscript: follow text with `_{text to be subscripted}`
* Example: \$x_{i + 1}\$ will be rendered as $x_{i + 1}$

Superscript: follow text with `^{text to be superscripted}`
* Example: \$x^{i + 1}\$ will be rendered as $x^{i + 1}$

Special characters: backslash and the name, which you probably will have to Google
* Example: \$\pi\$ will be rendered as $\pi$
* Example: \$\times\$ will be rendered as $\times$

Fractions: \frac{numerator}{denominator}
* Example: \$\frac{n(n + 1)}{2}\$ will be rendered as $\frac{n(n + 1)}{2}$

Square root: \$\sqrt{2}\$ will be rendered as $\sqrt{2}$

Sums: \$\sum_{n=1}^{\infty} 2^{-n} = 1\$ will be rendered as $\sum_{n=1}^{\infty} 2^{-n} = 1$

### Other Latex Functionality
Latex is capable of so much more! I recommend following a tutorial online to learn more. [ShareLatex](https://www.sharelatex.com/) is a great resource to learn more. For now, you know enough to typeset basic math formulas.

## Escaping Characters
Often characters/symbols have special meaning in Markdown. For example \$ denotes Latex math mode. Sometimes we really do want a \$ in our paragraphs! To do this, you can escape the functionality of a character by placing a backslash in front of the character. 

Example: `\\ \$` will render as \\ \$

Alternatively, I sometimes use inline code to escape built-in functionality.

Example: `\$\` will render as `$`
