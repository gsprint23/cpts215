## Navigating the File System
Your OS manages your file system. You can think of a file system as a tree structure, with the root node being the top-level folder (e.g. C:\ on Windows or / on Unix-based machines). Child nodes of the root node are folders/files in the top-level folder, and so on and so forth. Leaf nodes are files and empty folders.

We typically use a GUI such as Windows Explorer or Finder to access our files, manage our files, etc. All of the common tasks we use GUIs for have equivalent command line commands.

### Current Working Directory
When you launch the command prompt, the command prompt is operating in a folder (AKA a directory). The current folder is called the *current working directory*, or the CWD. In the command prompt, to the left of the cursor is the actual prompt, which denotes the CWD. You can also display the CWD with the following commands:

#### Windows
`cd`: displays the name of the current directory

```
C:\Users\gsprint>cd
C:\Users\gsprint
```

#### Unix-based
`pwd` (short for present working directory): print the full filename of the current working directory

```
gsprint@gsprint-x1:~/Documents/L1-2$ pwd
/home/gsprint/Documents/L1-2
```

### Paths
When you refer to a file or folder by name, the OS looks for it in the CWD.

If a file you want to interact with is in a directory other than the CWD, you will have to specify its path. The location of a file is represented by its path, the sequence of folders that the file is stored in, plus the file's name. There are two ways to specify a path:
1. Relative path: a path to a file or directory relative to the current directory.
1. Absolute path: a path to a file or directory specified by its exact location on your file system. An absolute path *uniquely* identifies the location of a file/folder on a file system.

#### Windows
On a windows machine, folders and file names in a path are separated by backslashes "\".

Relative path example: "files\transactions.txt" refers to a file ("transactions.txt") in a directory ("files") in the CWD.

Absolute path example:  "C:\Users\gsprint\cpts215\lessons\files\transactions.txt" refers to a file ("transactions.txt") in the folder "C:\Users\gsprint\cpts215\lessons\files" on my C:\ drive.

Note: Windows' absolute paths start with the drive, such as C:\, as the top-level directory.

Note: On a windows machine, folders and file names in a path are separated by backslashes "\". We know the backslash has a special purpose in Python, to escape certain characters, such as a newline "\n"; therefore, you will have to escape a backslash: "`\\`" in your path to a file: `"files\\transactions.txt"`. Alternatively, you can specify your path as a raw string: `r"files\transactions.txt"`. On a Unix-based machine (e.g. Mac, Linux distributions), the forward slash "/" is used in paths and you don't have to worry about this issue.

#### Unix-based
On a Unix-based machine, the forward slash "/" is used in paths.

Relative path example: "files/transactions.txt" refers to a file ("transactions.txt") in a directory ("files") in the CWD.

Absolute path example:  "/home/gsprint/cpts215/lessons/files/transactions.txt" refers to a file ("transactions.txt") in the folder "/home/gsprint/cpts215/lessons/files/transactions.txt" on my hard drive.

Note: Unix-based absolute paths start with root, /, as the top-level directory.

Note: The CWD can be referred to by a single period, `.`. The folder the CWD is in (i.e. CWD's parent folder) can be referred to by two periods, `..`. We will learn more about this when we talk about changing the CWD.

### Change the CWD
We can change the CWD by specifying a relative or absolute path to move to. 
1. Relative path
    1. `..`: Move "up" a folder (change the CWD to the parent folder, i.e. the folder containing the CWD)
    1. Move "down" a folder (change the CWD to a child folder, i.e. a sub folder in the CWD)
1. Absolute path
    1. Move to a folder specified by an absolute path
        
#### Windows
`cd <path>`: changes the current directory

```
C:\Users\gsprint\Documents\L1-2>cd
C:\Users\gsprint\Documents\L1-2

C:\Users\gsprint\Documents\L1-2>cd temp

C:\Users\gsprint\Documents\L1-2\temp>cd
C:\Users\gsprint\Documents\L1-2\temp
```

#### Unix-based
`cd <path>`: Change directory to `<path>`

```
gsprint@gsprint-x1:~$ pwd
/home/gsprint
gsprint@gsprint-x1:~$ cd Documents
gsprint@gsprint-x1:~$ pwd
/home/gsprint
```

## Set Permissions/Ownership of Files
Each file and folder has an owner (usually the creator) and certain user permission associated with it. We can view/modify the owner and permissions of files:

### Windows
`attrib`: Displays, sets, or removes attributes assigned to files or directories. If used without parameters, attrib displays attributes of all files in the current directory.

```
C:\Users\gsprint\Documents\L1-2\temp>attrib afile2.txt
A            C:\Users\gsprint\Documents\L1-2\temp\afile2.txt
```

### Unix-based
* `chmod`: Change permissions on a file
* `chown`: Change ownership of a file

Using the -l option of `ls` for a long listing. For example, `-rw-rw-r-- 1 gsprint gsprint   19 May  2 13:54 afile2.txt` says that (from right to left) afile2.txt was created on May 2 at 1:34pm and is of size 19 bytes. It belongs to the group `gsprint` and is owned by `gsprint`. It is one file and the file permissions are (from left to right):
* -: This is a normal file that contains any type of data (a directory would have a d here)
* r: The owner can read this file
* w: The owner can write to this file
* -: If there is an x here, the owner can execute this file
* r: The group can read this file
* w: The group can write to this file
* -: If there is an x here, the group can execute this file
* r: Any user can read this file
* -: If there is an w here, any user can write to this file
* -: If there is an x here, any user can execute this file

More about Linux file permissions can be read [here](https://www.linux.org/threads/file-permissions-chmod.4124/).

### Searching for a String in File(s)

#### Windows
* `find <search string> <path>`: searches for a text string in a file or files
* `find /i <search string> <path>`: ignore case
* `find /n <search string> <path>`: show line numbers
* `find /c <search string> <path>`: count number of lines
* `find /v <search string> <path>`: show lines that don't contain `<search string>`
* `findstr`: searches for strings in files (similar to `find` but has support for more features like [regular expressions](https://en.wikipedia.org/wiki/Regular_expression))

```
C:\Users\gsprint\Documents\L1-2\temp>find /i "TEXT" afile2.txt

---------- AFILE2.TXT
adding some text!!!
C:\Users\gsprint\Documents\L1-2\temp>find /i /n "TEXT" *.txt

---------- AFILE2.TXT
[1]adding some text!!!

---------- AFILE3.TXT
[3]line number 3 has TEXT!!
[5]line number 5 has more text...

---------- EMPTY.TXT
```

#### Unix-based
* `grep <search string> <file>`: search for `<search string>` in `<file>`
* `grep -r <search string> <file>`: recursive search in subdirectories
* `grep -i <search string> <file>`: ignore case

```
gsprint@gsprint-x1:~/Documents/L1-2$ ls
afile2.txt  afile.csv  temp
gsprint@gsprint-x1:~/Documents/L1-2$ grep -i "Text" *.txt
Adding some text!!
```

## Getting your IP/MAC Address
From [Wikipedia](https://en.wikipedia.org/wiki/IP_address):
>An IP address (abbreviation of Internet Protocol address) is an identifier assigned to each computer and other device (e.g., printer, router, mobile device, etc.) connected to a TCP/IP network[1] that is used to locate and identify the node in communications with other nodes on the network. IP addresses are usually written and displayed in human-readable notations, such as 172.16.254.1 in IPv4, and 2001:db8:0:1234:0:567:8:1 in IPv6.

From [Wikipedia](https://en.wikipedia.org/wiki/MAC_address):
>A media access control address (MAC address) of a computer is a unique identifier assigned to network interfaces for communications at the data link layer of a network segment. MAC addresses are used as a network address for most IEEE 802 network technologies, including Ethernet and Wi-Fi. Logically, MAC addresses are used in the media access control protocol sublayer of the OSI reference model. MAC addresses are most often assigned by the manufacturer of a network interface controller (NIC) and are stored in its hardware, such as the card's read-only memory or some other firmware mechanism.
