## `break` Statement
Sometimes we need to "break" out of a loop early, i.e. before the Boolean condition is False. We can accomlish this anywhere in the body of the loop statement with the `break` statement. Say, for example, when processing [words.txt](http://thinkpython2.com/code/words.txt), we want to print words in the file *until* we reach a special word the user has entered:


````python
def open_input_file(fname):
    '''
    
    '''
    in_file = open(fname, "r")
    return in_file

def close_file(file_2_close):
    '''
    
    '''
    file_2_close.close()
  
def display_words(in_file, stop_word):
    '''
    
    '''
    for line in in_file.readlines():
        line = line.strip()
        if line == stop_word:
            break
        print(line)

    
def main():
    '''
    
    '''
    fin = open_input_file("files\\words.txt")

    # try entering aardvark, it won't print that many words
    stop_word = input("Please enter the word to print words until: ")
    display_words(fin, stop_word)
        
    close_file(fin)
    
main()
````

## Nested Loops
Loops within loops! Just like how we can have `if` statements within `if` statements (nested `if` statements), we do the same with loops. We just need to be conscientious of:
* Indenting the bodies of the loops correctly
* Progress towards all of the Boolean conditions eventually being false


````python
for i in range(0, 5):
    for j in range(0, i):
        print("%d %d " %(i, j), end="")
    print("")
````

It can be helpful to keep track of the loop control variables with a table:

|i|j|output|
|-|-|-|
|0|0|\n|
|1|0|1 0 \n|
|2|0,1|2 0 2 1 \n|
|3|0,1,2|3 0 3 1 3 2 \n|
|4|0,1,2,3|4 0 4 1 4 2 4 3 \n|
|5||Exit loop|

Let's try it! Write a program to prompt the user to enter the number of lines of stars to print. On each line, there will be as many stars printed as the line number. For example, if the user enters 5, the first line will have 1 star, the second line will have 2 stars, etc., until the 5th line, which will have 5 stars:

```
* 
**
***
****
*****
```

````python
num_lines = int(input("Please enter the number of starry lines to print: "))

for i in range(0, num_lines):
    for j in range(0, i + 1):
        print("*", end="")
    print("")
````

Now try and print it the other way, with `num_lines` stars on the first line, `num_lines - 1` on the second, etc.

````python
num_lines = int(input("Please enter the number of starry lines to print: "))

for i in range(0, num_lines):
    for j in range(num_lines, i, -1):
        print("*", end="")
    print("")
````

## Guessing Game 2.0
Let's add functionality to the guessing game to allow a user to play the game as many times as they want. This will require an outer menu loop to keep the guessing game going until the user wants to quit, and an inner loop to guide the user to the correct answer. Before we write this code, let's take an aside and learn about random number generation. This is important because if the user plays our guessing game more than once, we want it to have a different number to guess. Hardly a game if the number is always 4!

### Random Numbers
To generate random numbers, we need to import the `random` module. Then, we will call the function `randrange(start_index, end_index)` to generate a random number in the range `start_index` to `end_index - 1`.

````python
import random

rand_num = random.randrange(0, 100)
print(rand_num)
````

Before generating random numbers, we often will *seed* the random number generating algorithm in order to get different random numbers on different runs of our programs. To learn more about random numbers and seeding, read the [documentation](https://docs.python.org/3/library/random.html).

````python
# calling seed() BEFORE generating random numbers will give different random numbers on subsequent runs of the program
random.seed()

rand_num = random.randrange(0, 100)
print(rand_num)
````

### Nested Loops and Random Numbers in Guessing Game
Now that we can use random numbers, let's do this.

````python
import random

def display_menu():
    '''
    
    '''
    print("**Welcome to my game!**")
    print("Please choose from the following options")
    print("1) View the game rules")
    print("2) Play the game")
    print("3) Quit")

def display_game_rules():
    '''
    
    '''
    print("The rules are fairly straightforward. Guess a number and I will tell you if you are correct, too high, or too low")
    
def play_guessing_game():
    '''
    
    '''
    num_to_guess = random.randrange(1, 11)
    players_guess = 0
    correct = False

    while not correct:
        print("Please enter a number between 1 and 10 inclusive")
        players_guess = int(input())

        # a guess is either equal to, greater than, or less than 
        if players_guess == num_to_guess: # BC 1
            print("Congrats, you guessed the number correctly")
            correct = True # exit loop
        elif players_guess > num_to_guess: # BC 2
            print("Your guess was too high")
        else: # players_guess < num_to_guess
            print("Your guess was too low")
    
    
def take_menu_action(choice):
    '''
    
    '''
    if choice == 1:
        display_game_rules()
    elif choice == 2:
        play_guessing_game()
    elif choice == 3:
        print("Thank you for playing")
    else:
        print("Not a valid menu option")
    
def main():
    '''
    
    '''
    choice = -1
    
    while choice != 3:
        display_menu()
        choice = int(input(">>"))
        take_menu_action(choice)

main()
````
